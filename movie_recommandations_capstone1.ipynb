```python
mean = movies['vote_average'].mean() #Total i.e the movies which r liked and disliked by users
```

```python
print(mean)
```

```python
 minvote = movies['num_voted_users'].quantile(0.8) #To get into the top movies list it should atleast get 957 votes (957 value is dependent on the value given inside the quantile [ps: quantile value should range only in between 0 to 1])
```

```python
print(minvote)
```

```python
 top_movies = movies.copy().loc[movies['num_voted_users'] >= minvote] #As we r copying the same data from the movies (.copy()) and .loc[] is used for accessing multiple columns. Here, we r taking the top movies by comparing with the every indiviual movie greater than the minvote which we have got.
```

```python
print(top_movies)
```

```python
top_movies.shape #so, there are 961 movies which have got more than 957 votes (top movies)
```

```python
# total minimum votes 
   # +
# contribution of individual voters 
  # =
# total average votes (for overall total score of that movie)

def weighted_avg_of_rating(x, minvote = minvote, mean = mean):
    voters = x['num_voted_users']
    avg_vote = x['vote_average']
    return (voters/(voters + minvote) * avg_vote) + (minvote/(minvote + voters) * mean)
# did only for one movie
```

```python
top_movies['score'] = top_movies.apply(weighted_avg_of_rating, axis =1) # for doing for all top movies #axis = 1 is for vertical (columns) and axis = 0 is for horizontal (rows)
```

```python
# to sort in a list like top 10 or top 20
top_movies = top_movies.sort_values('score',ascending = False) # ro set data in descending order we r using asencing = False # to set/sort from highest to lowest we are using ascending = False
```

```python
top_10 = top_movies[['movie_title','language','genres','num_voted_users','popularity','director_name','vote_average','score']]
```
